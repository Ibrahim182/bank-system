```python
class User:
    
    def __init__(self,Id,Password,name="",age=0,gender=""):
        self.__Id = Id
        self.__Password = Password
        self.__name = name
        self.__age = age
        self.__gender = gender
        
    # to change password in update information
    def set_pass(self):
        while True:
            try:
                new_pass = input('Enter your password **should contain number and character only** :')
                # to sure that password contain on;y number or character
                if re.findall('[a-zA-Z0-9]',new_pass):
                    self.__Password = new_pass
                    break
            except:
                print("Please enter valid password")
        
    # to set information when user sign up
    def set_info(self):

        while True:
            try:
                name = input('Enter name sir, : ')
                self.__name = name
                break
            except:
                print('Invalid input sir, :)')
                
        while True:
            try:
                age = int(input('Enter age sir, : '))
                self.__age = age
                break
            except:
                print('Invalid input sir, :)')
        while True:
            try:
                gender = input('Enter gender sir, : ')
                self.__gender = gender
                break
            except:
                print('Invalid input sir, :)')
        
    # all this function to return values of information to store it in database    
    def get_id(self):
        return self.__Id
        
    def get_pass(self):
        return self.__Password
        
    def get_name(self):
        return self.__name
    
    def get_age(self):
        return self.__age
    
    def get_gender(self):
        return self.__gender
    
    # to show details
    def show_det(self):
        print('--- Your inforamtion ---')
        print('name is ', self.__name)
        print('age is ',self.__age)
        print('Gender is ', self.__gender)
        

class Bank(User):
    def __init__(self,Id,Password,name="",age=0,gender=""):
        super().__init__(Id,Password,name,age,gender)
        self.__balance = 0
    
    # to set the value from database
    def set_bal(self,ind):
        self.__balance = df['Balance'][ind]
        
    # function deposit   
    def deposit(self,amount):
        self.amount = amount
        self.__balance = self.__balance + self.amount
        print(f'Your balance has been {self.balance} now')
        
    # function withdraw    
    def withdraw(self,amount):
        self.amount = amount
        # condition to sure valid withdraw
        if self.amount > self.__balance:
            print('Not enouc money')
        else:
            self.__balance = self.__balance - amount
            print(f'Your balance has been {self.balance} now')
            
    # to return balance to store it in database
    def get_bal(self):
        return self.__balance
    
    # view all details        
    def view(self):
        print('Balance is ', self.__balance)
```
